```python
import pandas as pd

# Sample data with NaN in 'status' (non-numeric) and missing temperature (numeric)
data = {
    'date': ['2020-05-06', '2020-05-09', '2020-05-10', '2020-05-11', '2020-05-12', '2020-05-13'],
    'temperature': [35.658200, 32.115275, 30.934300, 22.421250, 13.908200, 23.938200],
    'windSpeed': [10.788378, 8.449161, 7.669422, 6.889682, 19.012990, 19.012990],
    'status': ['sunny', None, 'rainy', 'cloudy', 'rainy', 'sunny']
}

# Convert to DataFrame and set 'date' as datetime index
df = pd.DataFrame(data)
df['date'] = pd.to_datetime(df['date'])  # Convert to datetime
df.set_index('date', inplace=True)       # Set as index

# Interpolate ONLY numeric columns (temperature, windSpeed)
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
df[numeric_cols] = df[numeric_cols].interpolate(method='time')

print(df)
```
